```python
def add_cm(height):
    return str(height) + 'cm'
```

```python
df['키'] = df['키'].apply(add_cm)
```

```python
def capitalize(lang): # sw특기 앞글자만 대문자로 변경
    if pd.notnull(lang):
        return lang.capitalize()
    return lang
```

```python
df['SW특기'] = df['SW특기'].apply(capitalize)
df
```
