### 퀴즈

```python
import pandas as pd
```

```python
df = pd.read_csv('./data/QuizData.csv')
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
X, y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
len(X_train), len(X_test)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
import matplotlib.pyplot as plt
```

####  -> 선형회귀 모델링

```python
plt.scatter(X_train, y_train, color='blue')
plt.show()
```

```python
plt.scatter(X_train, y_train, color='blue')
plt.plot(X_train, reg.predict(X_train), color='green')
plt.show()
```

```python
plt.scatter(X_test, y_test, color='blue')
plt.plot(X_train, reg.predict(X_train), color='green') #예측하는 선
plt.show()
```

```python
reg.score(X_train, y_train)
```

#### 다항회귀)

```python
df = pd.read_csv('./data/QuizData.csv') #오리지널 데이터 읽기
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2)
```

```python
import numpy as np
X_arange = np.arange(min(X), max(X), 0.1)
X_arange = X_arange.reshape(-1, 1) #2차배열로 변경
X_arange.shape
```

```python
X_poly = poly.fit_transform(X)
X_poly #X의 0승, X의 원래값, X * X값
```

```python
reg = LinearRegression()
reg.fit(X_poly, y)
```

```python
plt.figure(figsize=(8, 5), dpi=50)
plt.scatter(X, y, color='#6666FF') # 실제 
plt.plot(X_arange, reg.predict(poly.fit_transform(X_arange)), color='red') # 예측
plt.show()
```

```python
reg.score(X_poly, y)
```

```python
reg.predict(poly.fit_transform([[300]]))
```
